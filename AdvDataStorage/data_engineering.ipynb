# Cleaning Data for Analysis

```python
import pandas as pd
import numpy as np
```

```python
#csv files to read
measurecsvfile = "hawaii_measurements.csv"
stationcsvfile = 'hawaii_stations.csv'
```

```python
#read into dataframe
mdf = pd.read_csv(measurecsvfile)
mdf.head()
```

```python
#drop rows where prcp has NaN values
mdf = mdf.dropna(axis=0)
mdf.head()
```

```python
mdf.shape
```

```python
#reset index for dataframe after dropping rows
mdf = mdf.reset_index(drop=True)
mdf.head()
```

```python
# Save the cleaned data to a file
new_csv = "clean_hawaii_measurements.csv"
mdf.to_csv(new_csv, index=False)
```

```python
#read csv into dataframe
sdf = pd.read_csv(stationcsvfile, dtype=object)
sdf
```

```python
sdf.shape
```

```python
sdf = sdf.rename(index=str, columns={"station": "station_id"})
sdf.head()
```
